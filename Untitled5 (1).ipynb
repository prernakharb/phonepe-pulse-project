```python
#importing necessary packages

import pandas as pd
import json
import os
import psycopg2
import streamlit as st
import mysql.connector
```

```python
#to get aggregate transactions

path = '/content/pulse/data/aggregated/transaction/country/india/state/'
Aggregate_states = os.listdir(path)


columns={'States':[], 'Years':[],'Quarter':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}


for states in Aggregate_states:
    states_path = path + states + '/'
    Aggregate_year=os.listdir(states_path)

    for years in Aggregate_year:
        years_path =states_path + years + "/"
        file=os.listdir(years_path)

        for x in file:
            files_path = years_path + x
            data = open(files_path, 'r')
            A =json.load(data)

            try:
                for i in A['data']['transactionData'] :
                        name = i['name']
                        count=i['paymentInstruments'][0]['count']
                        amount=i['paymentInstruments'][0]['amount']
                        columns['Transacion_type'].append(name)
                        columns['Transacion_count'].append(count)
                        columns['Transacion_amount'].append(amount)
                        columns['States'].append(states)
                        columns['Years'].append(years)
                        columns['Quarter'].append(int(x.strip('.json')))
            except:
              pass
```

```python
agg_trans = pd.DataFrame(columns)
```

```python
agg_trans
```

```python
#to get aggregate users

path = '/content/pulse/data/aggregated/user/country/india/state/'
Aggregate_states = os.listdir(path)


user_columns={'States':[], 'Years':[],'Quarter':[],'user_brand':[], 'user_count':[], 'user_percentage':[]}


for states in Aggregate_states:
    states_path = path + states + '/'
    Aggregate_year=os.listdir(states_path)

    for years in Aggregate_year:
        years_path =states_path + years + "/"
        file=os.listdir(years_path)

        for x in file:
            files_path = years_path + x
            data = open(files_path, 'r')
            A =json.load(data)

            try:
                for i in A['data']['usersByDevice'] :
                        brand = i['brand']
                        count=i['count']
                        percentage=i['percentage']
                        user_columns['user_brand'].append(brand)
                        user_columns['user_count'].append(count)
                        user_columns['user_percentage'].append(percentage)
                        user_columns['States'].append(states)
                        user_columns['Years'].append(years)
                        user_columns['Quarter'].append(int(x.strip('.json')))
            except:
              pass
```

```python
agg_users = pd.DataFrame(user_columns)
```

```python
#to get map transactions

path3 = '/content/pulse/data/map/transaction/hover/country/india/state/'
map_states = os.listdir(path3)


mcolumn={'States':[], 'Years':[],'Quarter':[],'transaction_type':[], 'transaction_count':[], 'transaction_amount':[]}


for states in map_states:
    states_path = path3 + states + '/'
    map_year=os.listdir(states_path)

    for years in map_year:
        years_path =states_path + years + "/"
        file=os.listdir(years_path)

        for x in file:
            files_path = years_path + x
            data = open(files_path, 'r')
            A =json.load(data)

            try:
                for i in A['data']['hoverDataList'] :
                        name = i['name']
                        count=i['metric'][0]['count']
                        amount=i['metric'][0]['amount']
                        mcolumn['transaction_type'].append(name)
                        mcolumn['transaction_count'].append(count)
                        mcolumn['transaction_amount'].append(amount)
                        mcolumn['States'].append(states)
                        mcolumn['Years'].append(years)
                        mcolumn['Quarter'].append(int(x.strip('.json')))
            except:
              pass
```

```python
path4 = '/content/pulse/data/map/user/hover/country/india/state/'
map_users = os.listdir(path4)


mcolumn1={'States':[], 'Years':[],'Quarter':[],'Districts':[], 'registered_users':[], 'Appsopens':[]}


for states in map_users:
    states_path = path4 + states + '/'
    map_year=os.listdir(states_path)

    for years in map_year:
        years_path =states_path + years + "/"
        file=os.listdir(years_path)

        for x in file:
            files_path = years_path + x
            data = open(files_path, 'r')
            A =json.load(data)
            if ( A["data"]["hoverData"] == None):
                    pass
            else:

                try:
                    for i in A['data']['hoverData'].items() :
                            user_dist  = i[0]
                            user_reg =i[1]['registeredUsers']
                            user_app =i[1]['appOpens']
                            mcolumn1['Districts'].append(user_dist )
                            mcolumn1['registered_users'].append(user_reg )
                            mcolumn1['Appsopens'].append(user_app)
                            mcolumn1['States'].append(states)
                            mcolumn1['Years'].append(years)
                            mcolumn1['Quarter'].append(int(x.strip('.json')))
                except:
                  pass
```

```python
path5 = '/content/pulse/data/top/transaction/country/india/state/'
top_trans = os.listdir(path5)


columnsT={'States':[], 'Years':[],'Quarter':[],'District':[], 'count':[], 'Amount':[]}


for states in top_trans:
    states_path = path5 + states + '/'
    year_list =os.listdir(states_path)

    for years in year_list :
        years_path =states_path + years + "/"
        file=os.listdir(years_path)

        for x in file:
            files_path = years_path + x
            data = open(files_path, 'r')
            A =json.load(data)

            try:
                for i in A['data']['districts'] :
                        trans_district = i['entityName']
                        count=i['metric']['count']
                        amount=i['metric']['amount']
                        columnsT['District'].append(trans_district)
                        columnsT['count'].append(count)
                        columnsT['Amount'].append(amount)
                        columnsT['States'].append(states)
                        columnsT['Years'].append(years)
                        columnsT['Quarter'].append(int(x.strip('.json')))
            except:
              pass
```
